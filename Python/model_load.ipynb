```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
```

```python
from keras.models import load_model
loaded_model = load_model("my_model.keras")
```

```python
IMG_HEIGHT, IMG_WIDTH = 224, 224  # Replace with the input shape of your model if different

# Define a function to preprocess the image
def preprocess_image(image_path):
    img = load_img(image_path, target_size=(IMG_HEIGHT, IMG_WIDTH))
    img_array = img_to_array(img)  # Convert to NumPy array
    img_array = img_array / 255.0  # Normalize to [0, 1]
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    return img_array

# Function to classify an image
def classify_image(image_path, model):
    img_array = preprocess_image(image_path)
    predictions = model.predict(img_array)
    predicted_class = np.argmax(predictions, axis=1)[0]  # Get the class index
    confidence = np.max(predictions)  # Get the confidence score
    return predicted_class, confidence

# Classify the specific image
image_path = "images.jpg"  # Path to the image in your directory
predicted_class, confidence = classify_image(image_path, model)

# Output the result
print(f"Image: {image_path} -> Predicted Class: {predicted_class}, Confidence: {confidence:.2f}")
```
